```python
import tkinter as tk
from tkinter import messagebox

# Kitap verilerini saklamak için bir liste
kutuphane = []

# Kitap ekleme fonksiyonu
def kitap_ekle():
    kitap_adi = kitap_adi_entry.get()
    yazar_adi = yazar_adi_entry.get()
    
    if kitap_adi and yazar_adi:
        kutuphane.append({"isim": kitap_adi, "yazar": yazar_adi})
        kitap_listesi.insert(tk.END, f"{kitap_adi} - {yazar_adi}")
        kitap_adi_entry.delete(0, tk.END)
        yazar_adi_entry.delete(0, tk.END)
        messagebox.showinfo("Başarılı", "Kitap eklendi!")
    else:
        messagebox.showwarning("Hata", "Lütfen tüm alanları doldurun!")

# Kitap arama fonksiyonu
def kitap_ara():
    arama_kelimesi = arama_entry.get()
    sonuc_listesi.delete(0, tk.END)
    bulunan_kitaplar = [kitap for kitap in kutuphane if arama_kelimesi.lower() in kitap["isim"].lower() or arama_kelimesi.lower() in kitap["yazar"].lower()]
    
    if bulunan_kitaplar:
        for kitap in bulunan_kitaplar:
            sonuc_listesi.insert(tk.END, f"{kitap['isim']} - {kitap['yazar']}")
    else:
        messagebox.showinfo("Sonuç Yok", "Aradığınız kitap bulunamadı.")
    
# Tkinter ana penceresi
pencere = tk.Tk()
pencere.title("Kütüphane Uygulaması")

# Kitap Ekleme Alanı
kitap_adi_label = tk.Label(pencere, text="Kitap Adı")
kitap_adi_label.pack()

kitap_adi_entry = tk.Entry(pencere)
kitap_adi_entry.pack()

yazar_adi_label = tk.Label(pencere, text="Yazar Adı")
yazar_adi_label.pack()

yazar_adi_entry = tk.Entry(pencere)
yazar_adi_entry.pack()

kitap_ekle_buton = tk.Button(pencere, text="Kitap Ekle", command=kitap_ekle)
kitap_ekle_buton.pack()

# Kitap silme fonksiyonu
def kitap_sil():
    secilen_kitap = kitap_listesi.curselection()
    if secilen_kitap:
        kitap_listesi.delete(secilen_kitap)
        del kutuphane[secilen_kitap[0]]
        messagebox.showinfo("Başarılı", "Kitap silindi!")
    else:
        messagebox.showwarning("Hata", "Lütfen silmek için bir kitap seçin!")

# Kitap Silme Butonu
kitap_sil_buton = tk.Button(pencere, text="Kitap Sil", command=kitap_sil)
kitap_sil_buton.pack()


# Kitap Listesi
kitap_listesi_label = tk.Label(pencere, text="Kitap Listesi")
kitap_listesi_label.pack()

kitap_listesi = tk.Listbox(pencere)
kitap_listesi.pack()

# Kitap Arama Alanı
arama_label = tk.Label(pencere, text="Kitap Ara")
arama_label.pack()

arama_entry = tk.Entry(pencere)
arama_entry.pack()

arama_buton = tk.Button(pencere, text="Ara", command=kitap_ara)
arama_buton.pack()

# Arama Sonuçları
sonuc_listesi_label = tk.Label(pencere, text="Arama Sonuçları")
sonuc_listesi_label.pack()

sonuc_listesi = tk.Listbox(pencere)
sonuc_listesi.pack()

# Ana döngü
pencere.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk  # Modern görünümler için ttk kullanımı

# Kitap verilerini saklamak için bir liste
kutuphane = []

# Kitap ekleme fonksiyonu
def kitap_ekle():
    kitap_adi = kitap_adi_entry.get()
    yazar_adi = yazar_adi_entry.get()
    
    if kitap_adi and yazar_adi:
        kutuphane.append({"isim": kitap_adi, "yazar": yazar_adi})
        kitap_listesi.insert(tk.END, f"{kitap_adi} - {yazar_adi}")
        kitap_adi_entry.delete(0, tk.END)
        yazar_adi_entry.delete(0, tk.END)
        messagebox.showinfo("Başarılı", "Kitap eklendi!")
    else:
        messagebox.showwarning("Hata", "Lütfen tüm alanları doldurun!")

# Kitap arama fonksiyonu
def kitap_ara():
    arama_kelimesi = arama_entry.get()
    sonuc_listesi.delete(0, tk.END)
    bulunan_kitaplar = [kitap for kitap in kutuphane if arama_kelimesi.lower() in kitap["isim"].lower() or arama_kelimesi.lower() in kitap["yazar"].lower()]
    
    if bulunan_kitaplar:
        for kitap in bulunan_kitaplar:
            sonuc_listesi.insert(tk.END, f"{kitap['isim']} - {kitap['yazar']}")
    else:
        messagebox.showinfo("Sonuç Yok", "Aradığınız kitap bulunamadı.")

# Tkinter ana penceresi
pencere = tk.Tk()
pencere.title("Kütüphane Uygulaması")
pencere.geometry("500x600")
pencere.configure(bg="#e1f5fe")  # Arkaplan rengi (açık mavi)

# Başlık
baslik_label = tk.Label(pencere, text="📚 Kütüphane Uygulaması", font=("Helvetica", 16, "bold"), bg="#e1f5fe", fg="#0277bd")
baslik_label.pack(pady=10)

# Kitap Ekleme Alanı
frame_kitap_ekle = tk.Frame(pencere, bg="#e1f5fe")
frame_kitap_ekle.pack(pady=10)

kitap_adi_label = tk.Label(frame_kitap_ekle, text="Kitap Adı:", bg="#e1f5fe", fg="#01579b", font=("Arial", 12))
kitap_adi_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")

kitap_adi_entry = tk.Entry(frame_kitap_ekle, font=("Arial", 12), width=30)
kitap_adi_entry.grid(row=0, column=1, padx=5, pady=5)

yazar_adi_label = tk.Label(frame_kitap_ekle, text="Yazar Adı:", bg="#e1f5fe", fg="#01579b", font=("Arial", 12))
yazar_adi_label.grid(row=1, column=0, padx=5, pady=5, sticky="w")

yazar_adi_entry = tk.Entry(frame_kitap_ekle, font=("Arial", 12), width=30)
yazar_adi_entry.grid(row=1, column=1, padx=5, pady=5)

kitap_ekle_buton = tk.Button(frame_kitap_ekle, text="Kitap Ekle", font=("Arial", 12, "bold"), bg="#4caf50", fg="white", command=kitap_ekle)
kitap_ekle_buton.grid(row=2, column=0, columnspan=2, pady=10)

# Kitap Listesi
kitap_listesi_label = tk.Label(pencere, text="📖 Kitap Listesi", bg="#e1f5fe", fg="#0277bd", font=("Helvetica", 14, "bold"))
kitap_listesi_label.pack()

kitap_listesi = tk.Listbox(pencere, font=("Arial", 12), width=50, height=10, bg="#ffffff", fg="#000000", selectbackground="#bbdefb")
kitap_listesi.pack(pady=10)

# Kitap Arama Alanı
frame_arama = tk.Frame(pencere, bg="#e1f5fe")
frame_arama.pack(pady=10)

arama_label = tk.Label(frame_arama, text="Ara:", bg="#e1f5fe", fg="#01579b", font=("Arial", 12))
arama_label.grid(row=0, column=0, padx=5, pady=5, sticky="w")

arama_entry = tk.Entry(frame_arama, font=("Arial", 12), width=30)
arama_entry.grid(row=0, column=1, padx=5, pady=5)

arama_buton = tk.Button(frame_arama, text="Ara", font=("Arial", 12, "bold"), bg="#0288d1", fg="white", command=kitap_ara)
arama_buton.grid(row=0, column=2, padx=5, pady=5)

# Arama Sonuçları
sonuc_listesi_label = tk.Label(pencere, text="🔍 Arama Sonuçları", bg="#e1f5fe", fg="#0277bd", font=("Helvetica", 14, "bold"))
sonuc_listesi_label.pack()

sonuc_listesi = tk.Listbox(pencere, font=("Arial", 12), width=50, height=5, bg="#ffffff", fg="#000000", selectbackground="#bbdefb")
sonuc_listesi.pack(pady=10)

# Ana döngü
pencere.mainloop()
```
